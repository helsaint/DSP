```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def CosSignal(freq=440, amp=1.0, offset=0):
    period = 1/freq
    ts = np.linspace(0,3*period,101)
    signal = offset + amp*np.cos(2*np.pi*(freq)*ts)
    return signal, ts
```

```python
def SinSignal(freq=440, amp=1.0, offset=0):
    period = 1/freq
    ts = np.linspace(0,3*period,101)
    signal = offset + amp*np.sin(2*np.pi*(freq)*ts)
    return signal, ts
```

```python
cos_signal, x = CosSignal()
```

```python
plt.plot(x, cos_signal)
```

```python
sin_signal, x = SinSignal(freq=880, amp=0.5)
plt.plot(x, sin_signal)
```

```python
plt.plot(x, sin_signal+cos_signal+sin_signal)
```

```python
class Mammal:
    def __init__(self):
        self.type = "Mammal"
        self.blood = "Warm"
        self.vertebrate = True

    def print_mammal(self):
        print(self.type, self.blood)
        
class Person(Mammal):
    def __init__(self, name="John Does", height=10.1, weight=190):
        super().__init__()
        self.name = name
        self.height = height
        self.weight = weight

    def print_name(self):
        print("My name is: ", self.name)
    def print_height(self):
        print("My height is: ", self.height)
```

```python
class Signal:
    def __init__(self, func=np.cos, freq=440, amp=1.0, offset=0.0):
        self.func = func
        self.freq = freq
        self.amp = amp
        self.offset = offset
        self.period = 1/freq
        self.xs = np.asarray([])

    def _ys(self, ts):
        ys = self.offset + self.amp*self.func(2*np.pi*self.freq*ts)
        return ys

    def _dt(self):
        return 1/(5*2*np.pi*self.freq)

    def set_domain(self, domain=()):
        if not domain:
            self.xs = np.linspace(0,3*self.period, 101)
        else:
            dt = self._dt()
            num = int((domain[1] - domain[0])/dt)
            print(num)
            self.xs = np.linspace(domain[0], domain[1], num=num)
        
    def plot(self):
        if not (self.xs.any()):
            self.set_domain()
            ts = self.xs
        else:
            ts = self.xs
        ys = self._ys(ts)
        plt.plot(ts,ys)

    def properties(self):
        print("Frequency: ", self.freq)
        print("Period: ", self.period)
        print("Amplitude: ", self.amp)
        print("Offset: ", self.offset)
        print("Type: ", self.func)
```

```python
cos_signal = Signal()
cos_signal.plot()
```

```python
sin_signal = Signal(freq=880, func=np.sin, amp=0.5)
sin_signal.plot()
```

```python
class Number:
    def __init__(self, value=0.0):
        self.value = value
        
    def __add__(self, other):
        return Number(self.value + other.value)
    def get_value(self):
        return self.value
```

```python
class Vehicle:
    def __init__(self, started = False, speed = 0):
        self.started = started
        self.speed = speed
        print(self.speed)
    def start(self):
        self.started = True
        print("Started, let's ride!")
    def stop(self):
        self.speed = 0
    def increase_speed(self, delta):
        if self.started:
            self.speed = self.speed + delta
            print("Vrooooom! ", self.speed)
        else:
            print("You need to start me first")

class Car(Vehicle):
    trunk_open = False
    def open_trunk(self):
        print("trunk Open")
        self.trunk_open = True
    def close_trunk(self):
        self.trunk_open = False

class Motorcycle(Vehicle):
    def __init__(self, center_stand_out = False, speed=1):
        self.center_stand_out = center_stand_out
        super().__init__(self,speed)
    def start(self):
        self.started=True
        print("in Motorcycle.start()")

    def increase_speed(self,delta):
        self.speed = self.speed + delta
        print(self.speed)

moto = Motorcycle(speed=2)
print("####")
moto.start()
moto.increase_speed(10)
```

```python
import copy
'''
Signal is the parent class and will hold information
like amplitude, frequency, offset.
It also allows the user to directly input xs and ys 
values through the set_xs(), set_ys()
'''
class Signal:
    def __init__(self, freq=440.0, amp=1.0, offset=0.0):
        self.freq = freq
        self.amp = amp
        self.offset = offset
        self.xs = np.asarray([])
        self.ys = np.asarray([])

    def set_xs(self, xs=[]):
        self.xs = xs
    def set_ys(self, ys=[]):
        self.ys = ys

    def _dt(self):
        return 1/(5*2*np.pi*self.freq)

    def __add__(self,other):
        if(self.xs[-1] >= other.xs[-1]):
            new_signal = copy.copy(other)
            new_signal.xs = self.xs
            new_signal.set_ys()
            new_signal.ys = self.ys + new_signal.ys
        else:
            new_signal = copy.copy(self)
            new_signal.xs = other.xs
            new_signal.set_ys()
            new_signal.ys = self.ys + new_signal.ys
        return new_signal

    def plot(self):
        plt.plot(self.xs,self.ys)

'''
CosSignal and SinSignal will be used to do the actual construction
of sine and cosine waves.
'''
class CosSignal(Signal):
    def __init__(self, freq=440.0, amp=1.0, offset=0.0):
        super().__init__(freq=freq, amp=amp,offset=offset)
        self.set_xs()
        self.set_ys()

    def set_xs(self):
        period = 1/self.freq
        self.xs = np.linspace(0,3*period, 101)

    def set_ys(self):
        self.ys = self.offset + self.amp*np.cos(2*np.pi*self.freq*self.xs)

    def get_freq(self):
        return self.freq
    def get_amp(self):
        return self.amp
    def get_offset(self):
        return self.offset

class SinSignal(Signal):
    def __init__(self, freq=440.0, amp=1.0, offset=0.0):
        super().__init__(freq=freq, amp=amp,offset=offset)
        self.set_xs()
        self.set_ys()

    def set_xs(self):
        period = 1/self.freq
        self.xs = np.linspace(0,3*period, 101)

    def set_ys(self):
        self.ys = self.offset + self.amp*np.sin(2*np.pi*self.freq*self.xs)

    def get_freq(self):
        return self.freq
    def get_amp(self):
        return self.amp
    def get_offset(self):
        return self.offset

cos_signal_1 = CosSignal(freq=440.0)
sin_signal_2 = SinSignal(freq=880.0, amp=0.5)
sum_wave = cos_signal_1 + sin_signal_2
```
