```python
class Mammal:
    def __init__(self):
        self.type = "Mammal"
        self.blood = "Warm"
        self.vertebrate = True

    def print_mammal(self):
        print(self.type, self.blood)
        
class Person(Mammal):
    def __init__(self, name="John Does", height=10.1, weight=190):
        super().__init__()
        self.name = name
        self.height = height
        self.weight = weight

    def print_name(self):
        print("My name is: ", self.name)
    def print_height(self):
        print("My height is: ", self.height)
```

```python
person_1 = Person("Luke", "5 foot 6", 128)
person_2 = Person("Carrie", 100.5, "120 kg")
person_3 = Person()
```

```python
person_1.print_name()
person_1.print_height()
person_2.print_name()
person_2.print_height()
person_3.print_name()
person_3.print_height()
person_1.print_mammal()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Signal:
    def __init__(self, func=np.cos, freq=440, amp=1.0, offset=0.0):
        self.func = func
        self.freq = freq
        self.amp = amp
        self.offset = offset
        self.period = 1/freq
        self.xs = np.asarray([])

    def _ys(self, ts):
        ys = self.offset + self.amp*self.func(2*np.pi*self.freq*ts)
        return ys

    def _dt(self):
        return 1/(5*2*np.pi*self.freq)

    def set_domain(self, domain=()):
        if not domain:
            self.xs = np.linspace(0,3*self.period, 101)
        else:
            dt = self._dt()
            num = int((domain[1] - domain[0])/dt)
            print(num)
            self.xs = np.linspace(domain[0], domain[1], num=num)
        
    def plot(self):
        if not (self.xs.any()):
            self.set_domain()
            ts = self.xs
        else:
            ts = self.xs
        ys = self._ys(ts)
        plt.plot(ts,ys)

    def properties(self):
        print("Frequency: ", self.freq)
        print("Period: ", self.period)
        print("Amplitude: ", self.amp)
        print("Offset: ", self.offset)
        print("Type: ", self.func)
```

```python
cos_signal = Signal()
cos_signal.plot()
```

```python
sin_signal = Signal(freq=880, func=np.sin, amp=0.5)
sin_signal.plot()
```

```python
class Number:
    def __init__(self, value=0.0):
        self.value = value
        
    def __add__(self, other):
        return Number(self.value + other.value)
    def get_value(self):
        return self.value
```

```python
test1 = Number(value=3.5)
test2 = Number(value=4.5)
```

```python
t = test1 + test2
print(t, type(t), t.get_value())
```
